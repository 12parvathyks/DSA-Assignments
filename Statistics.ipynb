```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
datas=pd.read_csv('/content/Students_Performance.csv')
datas.head()
```

```python
datas['gender'].value_counts()
```

```python
datas['parental level of education'].value_counts().plot.bar()
plt.title('Comparison of Parental Education')
plt.xlabel('Qualification',color='red')
plt.ylabel('Count',color='red')
plt.show()
```

```python
datas["average score"] = datas["math score"] + datas["writing score"] + datas["reading score"] / 3
datas['average score']
```

```python
plt.title('Average score')
sns.barplot(x="gender", y="average score", data=datas)
```

```python
#seperate details of maths reading and writing scores
plt.title('Math Score')
sns.barplot(x="gender", y="math score", data=datas)
```

```python
plt.title('Reading Scores')
sns.barplot(x="gender", y="reading score", data=datas)
```

```python
plt.title('Writing Scores')
sns.barplot(x="gender", y="writing score", data=datas)
```

```python
#We can underatsnd that male score more in maths and females scored more both in reading and writing but the average score has more females.
```

```python
#1.c.ii)
plt.title('average score')
sns.barplot(hue="gender", y="average score", x="test preparation course", data=datas)
```

```python
plt.title('Math Scores')
sns.barplot(hue="gender", y="math score", x="test preparation course", data=datas)
```

```python
plt.title('Reading Scores')
sns.barplot(hue="gender", y="reading score", x="test preparation course", data=datas)
```

```python

plt.title('Writing Scores')
sns.barplot(hue="gender", y="writing score", x="test preparation course",data=datas)
```

```python
#So the students who completed the test preparation course scored higher in all the subjects.
```

```python
#1.d.i)
plt.title('Math Scores')
sns.barplot(x="race/ethnicity", y="math score", hue="test preparation course", data=datas)

```

```python
plt.title('Reading Scores')
sns.barplot(hue="test preparation course", y="reading score", x="race/ethnicity", data=datas)
```

```python
plt.title('Writing Scores')
sns.barplot(hue="test preparation course", y="writing score", x= 'race/ethnicity',data=datas)
```

```python
plt.title('Average Score')
sns.barplot(hue="test preparation course", y="average score", x= 'race/ethnicity',data=datas)
```
