```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Read the dataset to python environment.
iris_data=pd.read_excel('/content/iris.xls')
```

```python
#displaying the columns
iris_data.columns
```

```python
#Mean of each column
iris_data.mean()
```

```python
#Check for the null values present in the dataset.
iris_data.info()
```

```python
iris_data.isnull()
```

```python
#Visualizations
#Scatter Plot
#1
sns.scatterplot(data=iris_data)
```

```python
#Boxplot of Sepal width
#2
sns.boxplot(iris_data['SW'])
```

```python
#Bar chart of Classification and Petal length
#3
plt.bar(iris_data['Classification'],iris_data['PL'],color='Green')
plt.title('\nClassification on Petal length',size=15,color='red')
plt.xlabel("  \nClassification",size=12)
plt.ylabel("Petal length",size=12)
```

```python
#Seaborn Scatter plot
#4
plt.scatter(iris_data['PL'],iris_data['PW'],c=iris_data['SL'],marker='^',cmap='Greens')
```

```python
#scatter plot of Sepal width and and sepal length based on classification using Seaborn
#5
sns.scatterplot(x='SL', y='SW', hue='Classification', data=iris_data)
plt.show()
```

```python
#Pairplot
#6
sns.pairplot(iris_data,hue='Classification')
```

```python
### Box plot
#7
sns.boxplot(x='Classification',y='SL',data=iris_data)
```

```python
# Bar chart
#8
plt.bar(iris_data['SW'],iris_data['PW'],width=0.09)
plt.title('Sepal Width V/S Petal Width')
plt.xlabel('Sepal Width')
plt.ylabel('Petal Width')
```

```python
#Seaborn counplot
sns.countplot(x='Classification',data=iris_data)
```
